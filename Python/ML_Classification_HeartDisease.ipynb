# Data Preprocessing

Since there are not not any categorical data so we do not need to perform categorial encoding

```python
df
```

```python
x=df.iloc[:,1:-1].values
y=df.iloc[:,-1].values
x,y
```

#### Now lets split the dataset into training and testing dataset

```python
from sklearn.model_selection import train_test_split #importing train test split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

## Feature Scaling

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import confusion_matrix,classification_report
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train,x_test
```

# Now using two classifier Algorithms

## 1. Decision Tree

```python
model=RandomForestRegressor(n_estimators=160,random_state=6)
model.fit(x_train,y_train)
predicted=(model.predict(x_test))
predicted
print(accuracy_score(y_test, predicted.round())*100, "%")
```

```python
confusion_matrix(y_test, predicted.round())
```

```python
print(classification_report(y_test,predicted.round()))
```
