# Univariate Analysis

### Numerical Data

```python
sns.histplot(tips['total_bill'])
```

```python
sns.histplot(tips['total_bill'])
plt.title('Distribution of total bills')
```

```python
sns.histplot(tips['tip'])
plt.title('Distribution tips')
```

```python
sns.histplot(tips['size'])
plt.title('Distribution of size')
```

```python
#Boxplot which shows the mean line in central of box plot and sides line indicates the 25% AND 75% QUANTILE 
sns.boxplot(x=tips['tip'])
plt.title('Boxplot of Tips Amount')
```

```python
##here dots shows the outliers 
sns.boxplot(x=tips['size'])
plt.title('Boxplot of size')
```

```python
sns.boxplot(x=tips['total_bill'])
plt.title('Boxplot of total bills')
```

```python
#example of outlier 
ab=np.array([18, 19, 23, 24, 180])
ab.mean()
```

### Categorical Data

```python
sns.countplot(x='day',data=tips)
plt.title('Count observations per day')
```

```python
sns.countplot(x='sex',data=tips)
plt.title('Count of sex')
```

```python
sns.countplot(x='time',data=tips)
plt.title('Count of time')
```

```python
sns.countplot(x='smoker',data=tips)
plt.title('Count of smoker')
```

# Bivariate Analysis 

##### Numerical_numerical relationship 

```python
#scatterplot 
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='tip',palette='coolwarm')
plt.title('Total Bill vs Tip')
plt.show()
```

```python
sns.scatterplot(x='size',y='tip',data=tips,hue='tip',palette='coolwarm')
plt.title('size vs Tip')
plt.show()
```

```python
sns.scatterplot(x='total_bill',y='size',data=tips,hue='size',palette='coolwarm')
plt.title('Total Bill vs size')
plt.show()
```

```python
#correlation 
tips.corr(numeric_only=True)
```

```python
correlation=tips.corr(numeric_only=True)
sns.heatmap(correlation,annot=True,cmap='coolwarm')
plt.title('Correlation matrix')
```
