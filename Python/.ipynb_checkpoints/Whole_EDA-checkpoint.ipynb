# Feature Engineering 

```python
tips.head()
```

```python
#add new column for tip % of total bills
tips['tip_percent'] = (tips['tip']/tips['total_bill'])*100
```

```python
tips.head()
```

```python
#KDE for probability distribution plot (Karnel density estimation)
sns.histplot(tips['tip_percent'],kde=True)
plt.title('Histogram of tips percent')
```

```python
tips['tip_percent']. describe()
```

# Handling Outliers 

### Use domain knowledge

```python
# for outlier removing 
tips_adjusted=tips[tips['tip_percent']<40]
```

```python
tips_adjusted. head ()
```

```python
tips_adjusted['tip_percent']. describe()
```

### Using interquartile range

```python
#calculating quartiles
Q1 = tips['total_bill'].quantile(0.25)
Q3 = tips['total_bill'].quantile(0.75)
```

```python
Q1
```

```python
Q3
```

```python
#Interquartile range (IQR)
IQR = Q3 - Q1
print (IQR)
```

```python
#Boundary define 
lower_bound=Q1-1.5*IQR
upper_bound= Q3 + 1.5*IQR
```

```python
lower_bound
```

```python
upper_bound
```

```python
tips_clean=tips[(tips['total_bill']>=lower_bound) & tips['total_bill']<=upper_bound]
```

```python
tips_clean. head()
```

```python
tips_clean['total_bill']. describe()
```
