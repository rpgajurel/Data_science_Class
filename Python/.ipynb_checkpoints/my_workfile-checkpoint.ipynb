```python
##Let’s use matplotlib to visualise the series. Data visualisation
import matplotlib.pyplot as plt
```

```python
# Draw Plot
def plot_df(df, x, y, title="", xlabel='Year', ylabel='TR', dpi=100):
    plt.figure(figsize=(16,5), dpi=dpi)
    plt.plot(x, y, color='tab:red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()

plot_df(df, x=df.index, y=df.TR, title='Total Reveneu %') 
```
